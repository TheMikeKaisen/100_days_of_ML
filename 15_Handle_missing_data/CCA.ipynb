```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data_science_job.csv')
df.head()
```

```python
df.isnull().mean()*100
```

```python
df.shape
```

```python
cols = [var for var in df.columns if (df[var].isnull().mean()*100< 5 and df[var].isnull().mean()*100 >0) ]
cols
```

```python
df[cols].sample(5)
```

```python
new_df = df[cols].dropna()
```

```python
df.shape, new_df.shape
```

### compare the distribution of each column before and after imputation

```python
new_df.hist(bins=15, density=True, figsize=(12, 12))
plt.show()
```
